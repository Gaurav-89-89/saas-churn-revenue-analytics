```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("/saas_customer_churn_dataset.csv")


# 1. Overall churn rate

churn_rate = df["churn_flag"].mean()
print(f"Overall churn rate: {churn_rate:.2%}")


# 2. Churn rate by plan tier

plan_churn = df.groupby("plan_tier")["churn_flag"].mean() * 100
plan_churn.plot(kind="bar", title="Churn Rate by Plan Tier", ylabel="Churn Rate (%)")
plt.tight_layout()
plt.show()


# 3. Churn rate by contract type

contract_churn = df.groupby("contract_type")["churn_flag"].mean() * 100
contract_churn.plot(kind="bar", title="Churn Rate by Contract Type", ylabel="Churn Rate (%)")
plt.tight_layout()
plt.show()


# 4. Churn rate by company size

size_churn = df.groupby("company_size")["churn_flag"].mean() * 100
size_churn.plot(kind="bar", title="Churn Rate by Company Size", ylabel="Churn Rate (%)")
plt.tight_layout()
plt.show()


# 5. Feature adoption % vs churn (box plot)

df.boxplot(column="feature_adoption_pct", by="churn_flag")
plt.title("Feature Adoption % vs Churn")
plt.suptitle("")  # removes default pandas subtitle
plt.xlabel("Churned (0 = No, 1 = Yes)")
plt.ylabel("Feature Adoption %")
plt.tight_layout()
plt.show()


# 6. Support tickets vs churn rate

support_churn = df.groupby("support_tickets_last_90d")["churn_flag"].mean() * 100
support_churn.plot(kind="line", marker="o", title="Churn Rate by Support Tickets", ylabel="Churn Rate (%)")
plt.tight_layout()
plt.show()


# 7. Revenue lost to churn

total_mrr = df["mrr"].sum()
churned_mrr = df.loc[df["churn_flag"] == 1, "mrr"].sum()
print(f"Total MRR: ${total_mrr:,.2f}")
print(f"MRR lost to churn: ${churned_mrr:,.2f} ({churned_mrr/total_mrr:.2%})")


# 8. NPS score vs churn rate

nps_churn = df.groupby("nps_score")["churn_flag"].mean() * 100
nps_churn.plot(kind="bar", title="Churn Rate by NPS Score", ylabel="Churn Rate (%)")
plt.tight_layout()
plt.show()
```
